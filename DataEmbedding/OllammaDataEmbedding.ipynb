### Ollama Embedding Models

Embedding models are the models that are trained specifically to generate vector embeddings.
vector embeddings: long array of numbers that represent semantic meaning for a given specific text


```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
import os

# Fetch the environment variable and print its value
api_key = os.getenv("OPENAI_API_KEY")

# Check if the API key is None
if api_key is None:
    print("OPENAI_API_KEY is not set.")
else:
    # Set the environment variable
    os.environ["OPENAI_API_KEY"] = api_key

```

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = OllamaEmbeddings(model="gemma2:2b") ## By default it uses llama2
```

```python
embeddings
```

```python
emb_docs = embeddings.embed_documents(
    [
        "Anil is a good boy",
        "Anil likes to explore",
    ]
)
emb_docs
```

```python
emb_docs[1]
```

```python
len(emb_docs[0])
```

```python
query1 = embeddings.embed_query("What Anil likes?")
```

```python
query1
```
